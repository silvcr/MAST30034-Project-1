```python
import pandas as pd

train_data = pd.read_csv('../data/merged_data_2018.csv')
#creata an untipped trip column for the model
train_data['untipped_trips'] = train_data['count'] - train_data['tipped_trips']
```

```python
from sklearn.preprocessing import Normalizer

transformer = Normalizer()

train= transformer.fit_transform(train_data.drop(['tipped_trips', 'count', 'tip_rates', 
                                                        'taxi_zone', 'Unnamed: 0', 'untipped_trips'], axis = 1))
```

```python
import statsmodels.api as sm
from statsmodels.api import families
#fit the model
model = sm.GLM(endog = train_data[['tip_rates']], 
               exog = train, 
               family=families.Binomial())
binom = model.fit()
binom.summary()
```

```python
train_data
```

```python
#fitting the model with two parameters

train_new = transformer.fit_transform(train_data[['mean_trip_duration', 'mean_trip_distance']])
model_new = sm.GLM(endog = train_data[['tip_rates']], 
               exog = train_new, 
               family=families.Binomial())
binom_new = model_new.fit()
binom_new.summary()
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred = list(binom_new.predict())

y_true = list(train_data['tip_rates'])

print('r2 score: ', r2_score(y_true, y_pred))
print('mean_squared_error: ', mean_squared_error(y_true, y_pred))
```

## Processing the prediction data

```python
# Importing and starting a spark session
from pyspark.sql import SparkSession
from pyspark import SparkContext
#Supress warnings
spark = SparkSession.builder.getOrCreate()
sc = SparkContext.getOrCreate()
sc.setLogLevel('WARN')

spark.conf.set('spark.sql.execution.arrow.pyspark.enabled', True)
#Make the spark files present well
spark.conf.set('spark.sql.repl.eagerEval.enabled', True)
```

```python
#import taxi zones and nta data
import geopandas as gpd
taxi_zones = gpd.read_file('../raw_data_lite/taxi_zones.geojson')
tracts = gpd.read_file('../raw_data_lite/tracts.geojson')
```

```python
#merging the taxi zones and the NTA into a new dataframe
merged_zones = gpd.overlay(tracts, taxi_zones, how = 'intersection')

# filter out redundant columns
merged_zones = merged_zones.filter(['BoroCT2010', 'zone', 'LocationID', 'geometry'], axis = 1)
# create area column
merged_zones['area'] = merged_zones.geometry.area

#in cases where a tract overlaps more than one zone, remove all except the one with max overlap area.
merged_zones.sort_values(by='area', inplace=True)
merged_zones.drop_duplicates(subset='BoroCT2010', keep='last', inplace=True)

#create a dictionary for converting from tract code to taxi zone
tracts_to_zones = {}
for index, row in merged_zones.iterrows():
    tracts_to_zones[row['BoroCT2010']] = row['LocationID']
```

```python
taxi_2019 = spark.read.parquet('../preprocessed_data/preprocessed_taxi_2019.parquet')
```

```python
#loading in the data
from pyspark.sql.functions import *
census_2018 = pd.read_csv('../preprocessed_data/census_2018.csv')

#aggregating the taxi data to merge with census tracts via LocationID
taxi_2019 = taxi_2019.withColumn('tipped', (taxi_2019.tip_amount > 0).cast('int'))
economic_data = taxi_2019.groupBy('PULocationID') \
                                 .agg(mean('tipped').alias('tip_rates'),
                                      mean('trip_distance').alias('mean_trip_distance'),
                                      mean('trip_duration').alias('mean_trip_duration'),
                                      count('PULocationID').alias('count'),
                                      sum('tipped').alias('tipped_trips'))
economic_data = economic_data.toPandas()
economic_data
```

```python
taxi_zone_column = []
delete_index = []
error_counter = 0

#dictionary for converting counties to borough code:
boroughs = {'New': '1', 'Bronx' : '2', 
            'Kings': '3', 'Queens':'4', 
            'Richmond': '5'}
for index, row in census_2018.iterrows():
    # get the 6 digit tract code at the end
    code = row['GEO_ID'][-6:]
    
    #get the name of the county
    county = row['Name'].split()
    county = county[3]
    
    #convert the county to boro code and append to code
    code = boroughs[county] + code
    
    #handle missing values, as there is there is a difference in dataset lengths
    # upon extraction
    if code in tracts_to_zones.keys():
        taxi_zone_column.append(tracts_to_zones[code])
    else:
        delete_index.append(index)
```

```python
#delete non matching indices
census_2018 = census_2018.drop(index = delete_index)

#append taxi zones
census_2018['taxi_zone'] = taxi_zone_column
```

```python
census_2018 = census_2018.groupby('taxi_zone').agg('sum')
census_2018 = census_2018.reset_index()
census_2018
```

```python
predict_data = pd.merge(economic_data, census_2018, how = 'inner',
                         right_on = 'taxi_zone', left_on = 'PULocationID')
```

```python
#convert all the estimates into percentage estimates
income_estimates = ['Less than 10,000', '10,000 to 14,999', 
                    '15,000 to 24,999', '25,000 to 34,999',
                    '35,000 to 49,999', '50,000 to 74,999', 
                    '75,000 to 99,999', '100,000 to 149,999', 
                    '150,000 to 199,999', '200,000 or more']
#create a column for percentages, then drop its estimate counterpart
for income in income_estimates:
    predict_data[f'{income} percentage'] = (predict_data[f'{income} estimates'] 
                                           / predict_data['Total households estimates'])
    predict_data = predict_data.drop([f'{income} estimates'], axis = 1)

predict_data['Unemployment percentage'] = (predict_data['Unemployment estimates']
                                          / predict_data['Civilian labor force estimates'])
predict_data = predict_data.drop(['Unemployment estimates', 'Civilian labor force estimates',
                               'Total households estimates', 'Unnamed: 0'], axis = 1)
predict_data = predict_data.drop('PULocationID', axis = 1)
```

```python
import numpy as np
#fix cases where division resulted in NaN
predict_data = predict_data.replace(np.NaN, 0)
predict_data
```
